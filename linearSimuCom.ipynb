```python
def genR2(Y, D=3):
    probs = torch.zeros_like(Y)
    Ystd = Y - torch.mean(Y)
    probs = torch.exp(Ystd - D)
    probs = probs/(1 + probs)
    #probs[Ystd<0]
    uniSps = torch.rand_like(probs)
    R = uniSps <= probs
    R = R.type_as(Y).to_sparse()
    return R
```
